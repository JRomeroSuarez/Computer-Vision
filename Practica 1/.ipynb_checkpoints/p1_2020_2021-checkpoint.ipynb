###Â 1.1 Image creation

Remember that [Numpy](https://docs.scipy.org/doc/numpy-1.13.0/reference/) allows us to define images of different kinds treating them as matrices.

The most common examples are:

* Binary: with values of 0 or 1, interpreted as black or white, respectively.
* Grayscale: with possible data types uint8, uint16, double, whose pixel values define their intensity.
* RGB Image: color images consisting of 3 matrices (usually called 'channels'), which contain intensity values for each color separately R (red), G (green) and B (blue).


a) Create and display a grayscale image that has 150 rows and 200 columns where each value should be of type unsigned integer with 8 bits. 

```python
#Your solution
size = (150, 200) #Size of the image(Row, Columns)

dtype = np.uint8 #Type of the values

img = np.ndarray(size, dtype) #Grid with the size and type of values
cmap ="gray" #Colormap as gray

#Print the image.
plt.imshow(img, cmap)
plt.show()
```

How many values can each pixel have in this case? 

### Solution
This allows us to represent 2^8 = 256 values in each pixel.

b) Change the color of the pixels of the image in such a way that the left half is black (0) and the right half is white (255). 

Visualize the image and print the values of its first row.

```python
#Values of the height and width
height = img.shape[0]
width = img.shape[1]


img[:,:int(width/2)] = 0 #Paint from pixel 0 to middle width
img[:, int(width/2):] = 255 #Paint from middle width to last pixel


cmap ="gray" #Colormap as gray

#Print the image.
plt.imshow(img, cmap)
plt.show()
```

### 1.4 Images manipulation

a) Add a black frame to Lenna's image. It should have a width of 1/10th of the height of the original image. We recommend to create a bigger black image to later embed the original image in the central part.

```python
#Sizes of the image
img_height = lenna.shape[0]
img_width = lenna.shape[1]

#Type of the values
dtype = np.uint8 

#Margin of the image
margin = int(img_height *.1)

#Black image with the new margin
canvas = np.zeros((img_height+2*margin, img_width+2*margin,3),dtype)

#Embeding the original image with the black image
lenna_frame = canvas.copy()
lenna_frame[margin:canvas.shape[0] - margin, margin:canvas.shape[1] - margin] = lenna

plt.imshow(lenna_frame)
plt.show()
print("Dimensions:",canvas.shape)
```

b) Change the mask of the framed image to red. Visualize the three images with their respective titles.

```python
#Dimensions of the image
img_height = lenna.shape[0]
img_width = lenna.shape[1]
dtype = 'uint8' #Type of the values
margin = int(img_height *.1)

#Red color for paint the frame
red_color = np.array([255,0,0])


#Create the red frame
red_frame = np.full([img_height+2*margin, img_width+2*margin,3],red_color,dtype)

#Embed image in the middle of the black frame
lenna_redFrame = red_frame
lenna_redFrame[margin:canvas.shape[0] - margin, margin:canvas.shape[1] - margin] = lenna


plt.imshow(lenna_redFrame)
plt.show()

```

```python
fig=plt.figure(figsize=(15,15))
fig.add_subplot(1,3,1)
plt.imshow(lenna, cmap='gray')
plt.title('Lenna Original')
plt.xticks([])
plt.yticks([])

fig.add_subplot(1,3,2)
plt.imshow(lenna_frame, cmap='gray')
plt.title('Lenna Black Frame')
plt.xticks([])
plt.yticks([])

fig.add_subplot(1,3,3)
plt.imshow(lenna_redFrame, cmap='gray')
plt.title('Lenna Red Frame')
plt.xticks([])
plt.yticks([])

plt.show()        
```

###Â 1.5 Contrast enhancement

a) Open the image `car_lowContrast.jpg` and visualize it without graylevel enhancement. To this purpose, check which are the minimal and maximal values and how to use the command imshow without contrast enhancement.

Note: When running the function `imshow()` the image is authomatically enhanced. Use the `vmin` and `vmax` parameters in order to deactivate the enhancement applied by default.

```python
#Read the image
car_LW = io.imread('images/car_lowContrast.jpg')

#Show the image using the vmin and vmax values to show the image without contrast enchacement
plt.imshow(car_LW, vmin = 0, vmax = 255)
plt.show()

print("Max values: " + str(car_LW.max()))
print("Min values: " + str(car_LW.min()))
```

b) Create an image called `dark_image` using the  values from the original one, so that the minimal value of the new image is 0. Check the minimal and maximal values of the new image, and compare them with the previous one.

```python
#Decreasing the minimal value substracting
dark_image = car_LW - car_LW.min()

plt.imshow(dark_image, vmin = 0, vmax = 255)
plt.show()

print("Max values: " + str(dark_image.max()))
print("Min values: " + str(dark_image.min()))
```

c) Create a new image called `bright_im` from the original one, so that the maximal value of the new image is 255. Check the minimal and maximal values of the new image, and compare them with the previous one.

```python
#Maximal value adding the difference until 255
bright_im = car_LW + (255 - car_LW.max())

plt.imshow(bright_im, vmin = 0, vmax = 255)
plt.show()

print("Max values: " + str(bright_im.max()))
print("Min values: " + str(bright_im.min()))
```

d) Apply contrast enhancement to the original image. There are two ways for it: (1) by direct manipulation, and (2) through scikit-image command. Check the minimal and maximal values of the enhanced image, and compare them with the previous one. 

Implement the contrast enhancement using the two different available methods.

```python
#1st Method: Direct Manipulation
plt.imshow(car_LW, vmin = car_LW.min(), vmax = car_LW.max())
plt.show()

print("Max values: " + str(car_LW.max()))
print("Min values: " + str(car_LW.min()))
```

```python
#2nd Method: through scikit-image command
from skimage import exposure
car_LW2 = exposure.equalize_adapthist(car_LW)

plt.imshow(car_LW2)
plt.show()

print("Max values: " + str(car_LW.max()))
print("Min values: " + str(car_LW.min()))
```

### 1.6 Masks and logical image manipulation

Read the image `circles.bmp` (see figure top) and define 3 images in such a way that each one contains only 1 ball (see figure bottom). Visualize the 3 images in the same figure, each one with its corresponding title.

<img src="images_notebook/ex16.png">

**Note:** Be careful when applying `imshow` since it automatically increases the contrast of the image (putting minimum values to 0 and maximal to 1).

**Note:** Be careful when applying type conversion. Check the ranges of the image values.

```python
from skimage.color import rgb2gray
#Load image
circles = io.imread('images/circles.bmp')

#The circles were rgb and convert to grayscale
circles_gray = rgb2gray(circles)

#We take each color 
color = np.unique(circles_gray)

#Copy the image to work each positin
left_circle = circles_gray.copy()
central_circle = circles_gray.copy()
right_circle = circles_gray.copy()

circles = [left_circle,central_circle,right_circle]

for i in range(len(color)-1):
    circles[i][circles_gray != color[i]] = 1
    
    
fig=plt.figure(figsize=(15,15))
fig.add_subplot(1,3,1)
plt.imshow(left_circle, cmap='gray',vmin =0 , vmax = 1)
plt.title('left_circle')


fig.add_subplot(1,3,2)
plt.imshow(central_circle, cmap='gray',vmin = 0, vmax = 1)
plt.title('central_circle')


fig.add_subplot(1,3,3)
plt.imshow(right_circle, cmap='gray',vmin = 0, vmax = 1)
plt.title('right_circle')


plt.show()        
```

### 1.7 Color manipulation

a) Create and visualize the three images shown in the following figure:
<img src="./images_notebook/ex17a.png">

```python
#Size of the images
size = (50,50,3)

#Type of the values 
dtype = np.uint8

#Creating the figures with the size and dtype
img1 = np.zeros(size, dtype)
img2 = np.zeros(size, dtype)
img3 = np.zeros(size, dtype)

#Create the figures 
img1[:,img1.shape[1]//2:] = 255
img2[img2.shape[1]//2:,:] = 255
img3[:img3.shape[0]//2,:img3.shape[1]//2] = 255

#Print all the images 
fig=plt.figure(figsize=(15,15))
fig.add_subplot(1,3,1)
plt.imshow(img1, cmap='gray')
plt.title('img1')

fig.add_subplot(1,3,2)
plt.imshow(img2, cmap='gray')
plt.title('img2')

fig.add_subplot(1,3,3)
plt.imshow(img3, cmap='gray')
plt.title('img3')

plt.show()        
```

b) Using these three images, construct the following color image:
<img src='./images_notebook/ex17b.png'>

```python
#We search the white of the image1 and paint red(right)
img1[img1[...,0]== 255] = [255,0,0]
#We search the white of the image2 and paint green(downside), and we got yellow too
img2[img2[...,0]== 255] = [0,255,0]
#We search the white of the image3 and paint green(left upside)
img3[img3[...,0]== 255] = [0,0,255]

#We join all the matrices into one
img = img1 + img2 + img3

plt.imshow(img)
plt.show()
```

e) Given the images `coat.png` and `model.png` (first two images on the figure below), implement the necessary code to construct the image of the model with the coat (third image). **Note:** when applying resize, rotate, etc. keep track how the image type and values range change. When necessary, apply type conversion using img_as_float, img_as_uint, img_as_ubyte.

```python
#Read the coat and model images
coat = io.imread('images/coat.png')
model = io.imread('images/model.png')

#Sizes of the images
height = model.shape[0]
width = model.shape[1]

for i in range(height):
    for j in range(width):
        #We look at when it stops being black, that is, it is coat
        if(coat[i][j][0] != 0) and(coat[i][j][1] != 0) and (coat[i][j][2] != 0):
                model[i][j] = coat[i][j] #We change the pixels of the model for the coat
            
plt.imshow(model, cmap='gray')
```
