```python
# Your soluction here

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from skimage import img_as_float
from skimage.exposure import histogram

# Function for computing the difference of histogram of an image and the previous one
def compute_hist(im, im_prev, hist_im_prev=None):

    [hist_r, bins_r]=histogram((img_as_float(im)[:,:,0]),64)
    [hist_g, bins_g]=histogram((img_as_float(im)[:,:,1]),64)
    [hist_b, bins_b]=histogram((img_as_float(im)[:,:,2]),64)
    hist_im = np.array(hist_r+hist_g+hist_b)
        
    if hist_im_prev is None:
        [hist_r, bins_r]=histogram((img_as_float(im_prev)[:,:,0]),64)
        [hist_g, bins_g]=histogram((img_as_float(im_prev)[:,:,1]),64)
        [hist_b, bins_b]=histogram((img_as_float(im_prev)[:,:,2]),64)
        hist_im_prev = np.array(hist_r+hist_g+hist_b)

    hist_diff = np.mean(np.abs(hist_im_prev - hist_im))
        
    return hist_diff, hist_im_prev


# initialize the data arrays 
xdata = []
ydata = []
prev_hist = None

%matplotlib inline

for t in range(len(ic)-1):
    # update the data plot
    y, prev_hist = compute_hist(ic[t+1], ic[t])#, hist_im_prev=prev_hist)
    xdata.append(t) # time = x axis
    ydata.append(y) # y axis
    
fig, ax = plt.subplots(1)
plt.plot(xdata, ydata)
# We save this values to use them in a future exercise.
datax = xdata.copy()
datay = ydata.copy()
```

## 3.2 Background subtraction

Apply the background subtraction algorithm (check theory material).

a) Visualize the following images for each different scene (there are 3) of the video:

    1) images belonging to the shot
    2) the background image, and
    3) the foreground.
    
**Hint**: You can construct a mask obtained from the original image and the background in order to know which parts of the image form part from the foreground and recover from the original image just the foreground regions.

```python
# Your soluction here

# Create the plot figure to show the answers.
img, axarr = plt.subplots(nrows=len(shots),ncols=3,figsize=(10,15))
count = 0 # We'll use this variable to plot the images.
for scene in shots: # For each scene
    
    # Original image:
    # Add the original image to the figure.
    axarr[count,0].imshow(ic[scene[1]])
    axarr[count,0].set_title("1) Original Scene")
    
    # Background image 
    # Inicialize background as array of zeros and the sequence (to apply median filter) like a list of three lists 
    # (one for each channel RGB, for each channel we save the array of the images in this channel).
    background = np.zeros(ic[scene[1]].shape)
    sequence = [[], [], []]
    # Our sequence will be all the images between the frames that define the scene.
    for image in range(scene[0],scene[1]+1):
        # For each image between this two frames, we save the image array in each channel RGB (converted as float type 
        # with a values between 0 and 1).
        for channel in range(3):
            sequence[channel] += [img_as_float(ic[image][:, :, channel])]
    # Extract the background image using the median filter of each channel RGB.
    for i in range(3):
        background[:, :, i] = np.median(sequence[i], axis = 0)
    # Add the background image to the figure.
    axarr[count, 1].imshow(background)
    axarr[count,1].set_title("2) Background")
    
    # Foreground image:
    # The foreground corresponses to dynamic pixels of the sequence or, that is the same, substract the background 
    # (which corresponses to static pixels) from the original image.
    foreground = np.abs(img_as_float(ic[scene[1]]) - background)
    # Add the foreground image to the figure.
    axarr[count, 2].imshow(foreground)
    axarr[count,2].set_title("3) Foreground")
    count+=1

plt.show()
  
```

- What happens if the shots are not correctly extracted?
If the sequence which applies the median filter is wrong, the background and the foreground will be wrong, because they depend
of these shots.

- What happens if you find too many shots in the video?
The background, and therefore the foreground, won't be as precise as if we you more images. More information, carries more precision.

- What do the static background images represent?
It represents the pixels of the sequence that haven't changed, the static pixels.

- In which situations does the algorithm work and in which it does not?
If the sequence is equal to two frames or less, the algorithm doesn't work. Otherwise, it works.

- What happens if you substract the background image from the original one?
The result is the foreground.

- Do you see any additional application for this algorithm?
If you need to extract a channel rgb of a sequence.

k) Comment the differences between each method as well as their advantages and disadvantages.

The **Felzenszwalb algorithm**, works by defining a weighted graph in which weights are the amount of difference between pixels.

The **SLIC algorithm** works by defining an amount of clusters that we want to divide the original image in, calculating the average area of a cluster (total pixels / number of clusters), the average distance between cluster centers and iterating on these values until finding borders that fulfill these requisites.

The **Quickshift algorithm** works segmenting an RGB image(or any image with more than one channel) by identifying clusters of pixels in the joint spatial and color dimensions. Segments are local (superpixels) and can be used as a basis for further processing.
